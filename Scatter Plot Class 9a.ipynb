# Scatter Plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,40)
y = np.sin(x)
plt.plot(x,y, 'o' , color = "Darkorange")
plt.show()
```

```python
plt.plot(x,y, '-ok', color = 'red' )
plt.show()
```

```python
plt.plot(x,y, '-p', color='purple')
plt.show()
```

```python
x = np.random.randn(150)
y = np.random.randn(150)
colors = np.random.randn(150)

sizes = 1000 * np.random.randn(150)

plt.scatter(x,y, c=colors, s=sizes, alpha=0.3)
plt.show()
```

```python
y = np.random.randn(125)
z = np.random.randn(125)
colors = np.random.randn(125)

sizes = 1000 * np.random.randn(125)

plt.scatter(y,z, c=colors, s=sizes, alpha=0.4)
plt.show
```

```python
a = np.random.randn(130)
b = np.random.randn(130)
colors = np.random.randn(130)

sizes = 850 *  np.random.randn(130)


plt.scatter(a,b,  c=colors, s=sizes, alpha=0.25)
plt.colorbar()
plt.show()
```

```python
x = np.random.randn(1000)
plt.hist(x)
```

```python
b= np.random.randn(450)
plt.hist(b)
```

```python
e = np.random.randn(500)
plt.hist(e)
```

```python
plt.hist(e, bins=150, alpha=0.4, histtype='barstacked' , color='Orange')
plt.show()
```

```python
plt.hist(e, bins=150, alpha=0.4, histtype='stepfilled', color = 'DarkBlue')
plt.show()
```

```python
a1 = np.random.normal(0,0.9,1000)
a2 = np.random.normal(-3,2,1000)
a3 = np.random.normal(4,3,1000)


dict_extra = dict(histtype='stepfilled' , alpha=0.4, bins=25)

plt.hist(a1, **dict_extra)
plt.hist(a2, **dict_extra)
plt.hist(a3, **dict_extra)

plt.show()
```

```python
x= np.linspace(0,10,1000)
plt.plot(x , np.sin(x), '-b' , label="Sine Graph")
plt.plot(x , np.cos(x), '--r' , label="Cosine Graph")

plt.legend(loc='right' , ncol=2)
plt.show()
```

```python
x,y = np.random.multivariate_normal([0,0],[[1,1],[1,2]],10000).T
plt.hist2d(x,y, bins=15, cmap='Reds')
plt.show()
```

```python
a,b = np.random.multivariate_normal([0,1],[[2,2],[3,2]],10000).T
plt.hist2d(a,b , bins=25 , cmap='Blues')
plt.show()
```
